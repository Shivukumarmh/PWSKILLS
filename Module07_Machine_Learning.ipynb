4. What is the mathematical formula for Entropy?
  - It is negative of summation probability multiplied with the log of that probability with the base 2.
  
5. What is Information Gain, and how is it used in Decision Trees?
  - Information gain is the difference between impurity measure of parent and child.Information Gain quantifies the reduction in entropy (disorder) achieved by splitting a dataset on a specific feature.
  - It determines the best feature and split point for partitioning data.

6. What is the difference between Gini Impurity and Entropy?
  - Gini index has the square of probabilities whereas Entropy has the log of prbabilities with base 2.
  - Gini index is faster to compute that the Entropy.
  - Gini Index range [0,0.5]
  - Entropy range [0,1]

7. What is the mathematical explanation behind Decision Trees?
  - They work by splitting the dataset into subsets based on certain features, aiming to create subsets that are as homogeneous as possible with respect to the target variable.
  - It basically optimises split in the data to maximize homogeneity.

8. What is Pre-Pruning in Decision Trees?
  - Pre-pruning is when we stop the tree from growing too much by setting some conditions early on.
  - It prevents overfitting by limiting its growth before it becomes too complex.

9. What is Post-Pruning in Decision Trees?
  - It is basically cutting/pruning the tree once it is fully grown.
  - It is done in order to reduce the overfitting.

10. What is the difference between Pre-Pruning and Post-Pruning	?
  - Pre-Pruning: Initially the stopping conditions are set to prevent the growth of tree.
  - Post-Pruning : Once the tree is formed the tree is cut in terms of depth of tree.

11. What is a Decision Tree Regressor?
  - It is the decision tree when the target is continuous in nature.t operates by recursively splitting the dataset into subsets based on feature values, aiming to minimize the variance of the target variable within each subset.

12. What are the advantages and disadvantages of Decision Trees?
  - **Advantages:**
     - No Need for Data Preprocessing
     - Handles Mixed Data Types
     - Nonlinear Relationships and Feature Interactions
  - **Disadvantages:**
    - Overfitting
    - Bias with Imbalanced Data
    - Greedy Splitting

13.  How does a Decision Tree handle missing values?
  - It handles by:
    - Surrogate Splits
    - Weighted Decision Making
    - Some problem specifc solutions to handle the missing values.

14. How does a Decision Tree handle categorical features?
  - Ordinal Encoding Based on Target Association(Like One Hot Encoder.)
  - Binary Splits via Exhaustive Search
  -  Multi-way Splits

15. What are some real-world applications of Decision Trees?
  - Healthcare: Diagnosis and Treatment Optimization
  -  Finance: Risk Assessment and Fraud Detection
  -  Marketing: Customer Insights and Retention

```python
#16. Write a Python program to train a Decision Tree Classifier on the Iris dataset and print the model accuracy

from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

iris = datasets.load_iris()
X = iris.data
y = iris.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)

clf = DecisionTreeClassifier()
clf.fit(X_train, y_train)

y_pred = clf.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
#21. Write a Python program to train a Decision Tree Classifier with a maximum depth of 3 and compare its accuracy with a fully grown tree

from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

data = load_iris()
X = data.data
y = data.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)

clf_full = DecisionTreeClassifier()
clf_full.fit(X_train, y_train)

clf_partial= DecisionTreeClassifier(max_depth=3)
clf_partial.fit(X_train, y_train)

y_pred_full = clf_full.predict(X_test)
y_pred_partial = clf_partial.predict(X_test)

accuracy_full = accuracy_score(y_test, y_pred_full)
accuracy_partial = accuracy_score(y_test, y_pred_partial)

print("Accuracy with Full Tree:", accuracy_full)
print("Accuracy with Partial Tree:", accuracy_partial)
```

```python
#22.Write a Python program to train a Decision Tree Classifier using min_samples_split=5 and compare its accuracy with a default tree

#The min_samples_split=5 parameter in a Decision Tree controls the minimum number of samples required to
#split an internal node during training.

from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

data = load_iris()
X = data.data
y = data.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)

clf_default = DecisionTreeClassifier()
clf_default.fit(X_train, y_train)

clf_min_samples_split = DecisionTreeClassifier(min_samples_split=5)
clf_min_samples_split.fit(X_train, y_train)

y_pred_default = clf_default.predict(X_test)
y_pred_min_samples_split = clf_min_samples_split.predict(X_test)

accuracy_default = accuracy_score(y_test, y_pred_default)
accuracy_min_samples_split = accuracy_score(y_test, y_pred_min_samples_split)

print("Accuracy with Default Tree:", accuracy_default)
print("Accuracy with min_samples_split=5:", accuracy_min_samples_split)
```

```python
#23. Write a Python program to apply feature scaling before training a Decision Tree Classifier and compare its accuracy with unscaled data

from sklearn.tree import DecisionTreeClassifier
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler

data = load_iris()
X = data.data
y = data.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)

scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

dt_classifier = DecisionTreeClassifier()
dt_classifier.fit(X_train, y_train)
y_pred = dt_classifier.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)

dt_scaled_classifier = DecisionTreeClassifier()
dt_scaled_classifier.fit(X_train_scaled, y_train)
y_pred_scaled = dt_scaled_classifier.predict(X_test_scaled)
accuracy_scaled = accuracy_score(y_test, y_pred_scaled)

print("Accuracy with unscaled data:", accuracy)
print("Accuracy with scaled data:", accuracy_scaled)
```

```python
#26. Write a Python program to train a Decision Tree Regressor with max_depth=5 and compare its performance with an unrestricted tree

from sklearn.datasets import fetch_california_housing
from sklearn.tree import DecisionTreeRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

california = fetch_california_housing()
X = california.data
y = california.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)

regressor_full = DecisionTreeRegressor()
regressor_full.fit(X_train, y_train)

regressor_limited = DecisionTreeRegressor(max_depth=5)
regressor_limited.fit(X_train, y_train)

y_pred_full = regressor_full.predict(X_test)
y_pred_limited = regressor_limited.predict(X_test)

mse_full = mean_squared_error(y_test, y_pred_full)
mse_limited = mean_squared_error(y_test, y_pred_limited)

print("MSE with Full Tree:", mse_full)
print("MSE with Limited Tree:", mse_limited)

```

```python
#27. Write a Python program to train a Decision Tree Classifier, apply Cost Complexity Pruning (CCP), and visualize its effect on accuracy

import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

iris = load_iris()
X = iris.data
y = iris.target

clf = DecisionTreeClassifier()
clf.fit(X, y)

path = clf.cost_complexity_pruning_path(X, y)
ccp_alphas, impurities = path.ccp_alphas, path.impurities

clfs = []
for ccp_alpha in ccp_alphas:
    clf = DecisionTreeClassifier(random_state=0, ccp_alpha=ccp_alpha)
    clf.fit(X, y)
    clfs.append(clf)

train_scores = [clf.score(X, y) for clf in clfs]
test_scores = [clf.score(X, y) for clf in clfs]

fig, ax = plt.subplots()

ax.plot(ccp_alphas, train_scores, marker='o', label='Train', drawstyle='steps-post')
ax.plot(ccp_alphas, test_scores, marker='o', label='Test', drawstyle='steps-post')

```

```python
#28.  Write a Python program to train a Decision Tree Classifier and evaluate its performance using Precision,Recall, and F1-Score

from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import precision_score, recall_score, f1_score

iris = load_iris()
X = iris.data
y = iris.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)

clf = DecisionTreeClassifier()
clf.fit(X_train, y_train)

y_pred = clf.predict(X_test)

precision = precision_score(y_test, y_pred, average='weighted')
recall = recall_score(y_test, y_pred, average='weighted')
f1 = f1_score(y_test, y_pred, average='weighted')

print("Precision:", precision)
print("Recall:", recall)
print("F1-Score:", f1)
```

```python
#29. Write a Python program to train a Decision Tree Classifier and visualize the confusion matrix using seaborn

import seaborn as sns
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot

iris = load_iris()

X = iris.data
y = iris.target

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)

model=DecisionTreeClassifier()
model.fit(X_train, y_train)

y_pred = model.predict(X_test)

cm = confusion_matrix(y_test, y_pred)

sns.heatmap(cm, annot=True)

matplotlib.pyplot.show()


```
