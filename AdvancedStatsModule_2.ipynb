##    Theory

1. What is a random variable in probability theory?
  - In probability theory, a random variable is a function that assigns numerical values to the outcomes of a random experiment. It provides a way to quantify uncertainty by mapping outcomes from a sample space to real numbers.

2. What are the types of random variables?
  -  Discrete Random Variable
  -  Continuous Random Variable

3.  What is the difference between discrete and continuous distributions?
  - Discrete Distributions: It takes only the whole number as the assigned value on the number line.
  - Continuous Distribution: It takes every number on the number line as the value.

4.  What are probability distribution functions (PDF)?
  - A Probability Distribution Function (PDF) refers to a function that describes the likelihood of different outcomes of a random variable. The type of PDF depends on whether the random variable is discrete or continuous.

5.  How do cumulative distribution functions (CDF) differ from probability distribution functions (PDF)?
  - CDF gives the summation of value till that point whereas the PDF give the value of that point only.

6.  What is a discrete uniform distribution ?
  - A Discrete Uniform Distribution is a probability distribution where a finite number of outcomes are equally likely. Each possible value of the random variable has the same probability.

7.  What are the key properties of a Bernoulli distribution?
  - It has only 2 possible outcomes one is success with the probability p and the other is failure with probability 1-p.

8.  What is the binomial distribution, and how is it used in probability?
  - The Binomial Distribution is a discrete probability distribution that models the number of successes in a fixed number of independent Bernoulli trials, where each trial has only two possible outcomes.

9.  What is the Poisson distribution and where is it applied?
  - The Poisson distribution is a discrete probability distribution that models the number of times an event occurs in a fixed interval of time or space, given that the events occur independently and at a constant average rate.

10.  What is a continuous uniform distribution?
  - The continuous uniform distribution is a probability distribution in which all outcomes within a given range are equally likely. It is defined over a continuous interval [a,b], meaning that any value in this range has an equal probability of occurring.

11.  What are the characteristics of a normal distribution?
  - The normal distribution (also called the Gaussian distribution) is a continuous probability distribution that is symmetric around its mean. It is one of the most important probability distributions in statistics and machine learning.

12. What is the standard normal distribution, and why is it important?
  - It normal distribution with the mean = 0 and standard deviation = 1. It is important as it forms a common ground to compare various values.

13. What is the Central Limit Theorem (CLT), and why is it critical in statistics?
  - When we take a large number of random samples from any population (regardless of its original distribution), the distribution of the sample means will approach a normal distribution as the sample size increases.
  - Basically the sample mean follows a normal distribution as n increases.

14.  How does the Central Limit Theorem relate to the normal distribution?
  - CLT shows that, regardless of the original populationâ€™s distribution, the sampling distribution of the sample mean will be approximately normal if the sample size is sufficiently large.

15. What is the application of Z statistics in hypothesis testing?
  - It  is used in hypothesis testing when we want to determine whether a sample mean significantly differs from a known population mean.
  - Used when:
      - When we know the standard deviation of the population.
      - When the sample size is greater than 30,ensuring normality by the Central Limit Theorem.

16. How do you calculate a Z-score, and what does it represent?
  - It measures how far a data point is from the mean, in terms of standard deviations. It tells us how unusual or typical a value is within a dataset.

17. What are point estimates and interval estimates in statistics?
  - A point estimate is a single value used to approximate a population parameter. It provides an exact estimate based on sample data.
  - An interval estimate provides a range of values within which the population parameter is likely to lie. It is more reliable than a point estimate because it accounts for variability and uncertainty.

18.  What is the significance of confidence intervals in statistical analysis?
  - It increases the confidence of the estimate.

19. What is the relationship between a Z-score and a confidence interval?
  - The Z-score determines the confidence level of an interval estimate.
  - A larger Z-score leads to a wider confidence interval.

20.  How are Z-scores used to compare different distributions?
  - It is used to standardize values from different distributions, allowing for meaningful comparisons. By converting raw values into Z-scores, we can compare data from distributions with different means and standard deviations.

21. What are the assumptions for applying the Central Limit Theorem?
  - Sample Size Should Be Large
  - Finite Variance
  - Random Sampling and Independent Observations.

22. What is the concept of expected value in a probability distribution?
  - It is the mean of the sample for a given population.It represents the theoretical mean of the distribution.

23. How does a probability distribution relate to the expected outcome of a random variable?
  - The expected value E(X) is calculated using the probabilities and possible values of a random variable.

```python
#3. Write a Python function to calculate the probability distribution function (PDF) of a Bernoulli distribution
def bernoulli_pmf(x, p):
    if x not in [0, 1]:
        raise ValueError("x must be 0 or 1 for a Bernoulli distribution.")
    if not (0 <= p <= 1):
        raise ValueError("p must be between 0 and 1.")

    return (p ** x) * ((1 - p) ** (1 - x))

# Example usage
p = 0.7  # Probability of success
print(f"P(X=0) = {bernoulli_pmf(0, p)}")
print(f"P(X=1) = {bernoulli_pmf(1, p)}")
```

```python
#9.  Write a Python function to calculate Z-scores from a dataset and plot them

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

def calculate_and_plot_zscores(data):
    mean = np.mean(data)
    std_dev = np.std(data, ddof=1)  # Using sample standard deviation (ddof=1)

    z_scores = (data - mean) / std_dev

    sns.histplot(z_scores, bins=30, kde=True, color='blue', edgecolor='black', alpha=0.7)

    # Labels and title
    plt.xlabel('Z-Score')
    plt.ylabel('Frequency')
    plt.title('Z-Score Distribution')


    # Show plot
    plt.show()

    return z_scores

sample_data = np.random.normal(loc=50, scale=10, size=1000)  # Mean=50, Std=10
z_scores = calculate_and_plot_zscores(sample_data)
```

```python
#11.  Simulate multiple samples from a normal distribution and verify the Central Limit Theorem
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


mu, sigma = 50, 10  # Mean and Standard Deviation
population = np.random.normal(mu, sigma, 10000)  # Large normal population

# Step 2: Define sample sizes
sample_sizes = [5, 30, 100]  # Small, medium, and large sample sizes
num_samples = 1000  # Number of samples

# Step 3: Compute sample means for each sample size
sample_means = {
    n: [np.mean(np.random.choice(population, size=n)) for _ in range(num_samples)]
    for n in sample_sizes
}

# Step 4: Plot the original normal distribution
plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
sns.histplot(population, bins=50, kde=True, color='blue', alpha=0.7, edgecolor='black')
plt.title('Original Normal Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Step 5: Plot the sampling distributions for different sample sizes
for i, n in enumerate(sample_sizes, start=2):
    plt.subplot(2, 2, i)
    sns.histplot(sample_means[n], bins=30, kde=True, color='green', alpha=0.7, edgecolor='black')
    plt.title(f'Sampling Distribution (n={n})')
    plt.xlabel('Sample Mean')
    plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
#14. Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal distribution
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

def calculate_z_score(data, x):

    mean = np.mean(data)
    std_dev = np.std(data, ddof=1)
    z_score = (x - mean) / std_dev
    return z_score


data = np.random.normal(loc=50, scale=10, size=1000)  # Mean = 50, Std Dev = 10, 1000 data points


x_value = 65  # Example
z_value = calculate_z_score(data, x_value)
print(f"Data point: {x_value}, Z-score: {z_value:.2f}")


x = np.linspace(-4, 4, 1000)
y = stats.norm.pdf(x, 0, 1)

plt.figure(figsize=(8, 5))
plt.plot(x, y, label='Standard Normal Distribution', color='blue')

# Mark the Z-score position
plt.axvline(z_value, color='red', linestyle='dashed', label=f'Z-score = {z_value:.2f}')


x_fill = np.linspace(-4, z_value, 500)
y_fill = stats.norm.pdf(x_fill, 0, 1)
plt.fill_between(x_fill, y_fill, color='green', alpha=0.3)

# Labels and legend
plt.xlabel('Z-Score')
plt.ylabel('Probability Density')
plt.title('Z-Score Comparison with Standard Normal Distribution')
plt.legend()
plt.grid(True)

plt.show()
```

```python
#15. 5 Implement hypothesis testing using Z-statistics for a sample dataset
import numpy as np
import scipy.stats as stats

def z_test(sample, pop_mean, pop_std, alpha=0.05):

    n = len(sample)
    sample_mean = np.mean(sample)
    se = pop_std / np.sqrt(n)


    z_score = (sample_mean - pop_mean) / se

    # Compute p-value for a two-tailed test
    p_value = 2 * (1 - stats.norm.cdf(abs(z_score)))


    reject_null = p_value < alpha

    return z_score, p_value, reject_null



sample_data = np.random.normal(loc=172, scale=6, size=50)  # Mean=172, Std=6, Sample size=50


population_mean = 170  # Null hypothesis mean
population_std = 6  # Known population standard deviation

# Step 3: Perform Z-test
z_score, p_value, reject_null = z_test(sample_data, population_mean, population_std)

# Step 4: Display results
print(f"Z-Score: {z_score:.3f}")
print(f"P-Value: {p_value:.5f}")
if reject_null:
    print("Conclusion: Reject the null hypothesis (H0).")
else:
    print("Conclusion: Fail to reject the null hypothesis (H0).")
```

```python
#16. Create a confidence interval for a dataset using Python and interpret the result
import numpy as np
import scipy.stats as stats

def confidence_interval(data, confidence=0.95):

    n = len(data)
    sample_mean = np.mean(data)
    sample_std = np.std(data, ddof=1)


    se = sample_std / np.sqrt(n)
    z_score = stats.norm.ppf(1 - (1 - confidence) / 2)

    # Compute confidence interval bounds
    margin_of_error = z_score * se
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    return lower_bound, upper_bound, sample_mean



sample_data = np.random.normal(loc=75, scale=10, size=50)  # Mean=75, Std=10, Sample size=50


lower, upper, mean = confidence_interval(sample_data, confidence=0.95)


print(f"Sample Mean: {mean:.2f}")
print(f"95% Confidence Interval: ({lower:.2f}, {upper:.2f})")
```

```python
#17. Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def confidence_interval(data, confidence=0.95):

    n = len(data)
    sample_mean = np.mean(data)
    sample_std = np.std(data, ddof=1)


    se = sample_std / np.sqrt(n)


    z_score = stats.norm.ppf(1 - (1 - confidence) / 2)


    margin_of_error = z_score * se
    lower_bound = sample_mean - margin_of_error
    upper_bound = sample_mean + margin_of_error

    return lower_bound, upper_bound, sample_mean



sample_data = np.random.normal(loc=100, scale=15, size=50)
lower, upper, mean = confidence_interval(sample_data, confidence=0.95)


print(f"Sample Mean: {mean:.2f}")
print(f"95% Confidence Interval: ({lower:.2f}, {upper:.2f})")


plt.figure(figsize=(8, 5))
plt.hist(sample_data, bins=10, color='skyblue', edgecolor='black', alpha=0.7, density=True, label='Sample Data')

# Mark mean and confidence interval bounds
plt.axvline(mean, color='red', linestyle='dashed', linewidth=2, label=f'Mean = {mean:.2f}')
plt.axvline(lower, color='green', linestyle='dashed', linewidth=2, label=f'Lower CI = {lower:.2f}')
plt.axvline(upper, color='green', linestyle='dashed', linewidth=2, label=f'Upper CI = {upper:.2f}')

# Labels and legend
plt.xlabel('Value')
plt.ylabel('Density')
plt.title('Histogram of Sample Data with 95% Confidence Interval')
plt.legend()
plt.grid(True)

plt.show()
```

```python
#22.  Calculate the range and interquartile range (IQR) of a dataset generated from a normal distribution
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

np.random.seed(42)
data = np.random.normal(loc=50, scale=15, size=1000)


data_range = np.max(data) - np.min(data)


Q1 = np.percentile(data, 25)
Q3 = np.percentile(data, 75)
IQR = Q3 - Q1


plt.figure(figsize=(8, 5))
sns.histplot(data, bins=30, kde=True, color='skyblue', label="Data Distribution")


plt.axvline(Q1, color='green', linestyle='dashed', linewidth=2, label=f'Q1 (25th) = {Q1:.2f}')
plt.axvline(Q3, color='red', linestyle='dashed', linewidth=2, label=f'Q3 (75th) = {Q3:.2f}')
plt.axvline(np.min(data), color='black', linestyle='dashed', linewidth=1, label=f'Min = {np.min(data):.2f}')
plt.axvline(np.max(data), color='black', linestyle='dashed', linewidth=1, label=f'Max = {np.max(data):.2f}')


plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Normal Distribution with Range and IQR")
plt.legend()
plt.grid(True)


plt.show()


print(f"Range: {data_range:.2f}")
print(f"Interquartile Range (IQR): {IQR:.2f}")
```

```python
#23. Implement Z-score normalization on a dataset and visualize its transformation

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


data = np.random.normal(loc=50, scale=15, size=1000)


mean = np.mean(data)
std_dev = np.std(data, ddof=1)


z_scores = (data - mean) / std_dev


plt.figure(figsize=(10, 5))


plt.subplot(1, 2, 1)
sns.histplot(data, bins=30, kde=True, color='skyblue', label="Original Data")
plt.axvline(mean, color='red', linestyle='dashed', linewidth=2, label=f'Mean = {mean:.2f}')
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Original Distribution")
plt.legend()
plt.grid(True)


plt.subplot(1, 2, 2)
sns.histplot(z_scores, bins=30, kde=True, color='orange', label="Z-Score Normalized Data")
plt.axvline(0, color='red', linestyle='dashed', linewidth=2, label='Mean = 0')
plt.xlabel("Z-Score")
plt.ylabel("Frequency")
plt.title("Z-Score Normalized Distribution")
plt.legend()
plt.grid(True)


plt.tight_layout()
plt.show()


print(f"Original Mean: {mean:.2f}, Original Std Dev: {std_dev:.2f}")
print(f"Normalized Mean: {np.mean(z_scores):.2f}, Normalized Std Dev: {np.std(z_scores, ddof=1):.2f}")
```

```python
#24. Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal distribution.

import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns

def analyze_distribution(data):

    skewness = stats.skew(data)
    kurtosis = stats.kurtosis(data, fisher=True)


    plt.figure(figsize=(8, 5))
    sns.histplot(data, bins=30, kde=True, color='skyblue', label="Data Distribution")


    plt.xlabel("Value")
    plt.ylabel("Frequency")
    plt.title(f"Skewness: {skewness:.2f}, Kurtosis: {kurtosis:.2f}")
    plt.axvline(np.mean(data), color='red', linestyle='dashed', linewidth=2, label=f'Mean = {np.mean(data):.2f}')
    plt.legend()
    plt.grid(True)


    plt.show()


    return skewness, kurtosis


np.random.seed(42)
data = np.random.normal(loc=50, scale=15, size=1000)


skew, kurt = analyze_distribution(data)


print(f"Skewness: {skew:.2f}")
print(f"Kurtosis (Excess): {kurt:.2f}")
```
